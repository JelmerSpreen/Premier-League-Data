```python
import pandas as pd
df = pd.read_excel(r'C:\Users\Jelmer\Documents\Data science project Premier Leauge\Premier_clubs_list_final.xlsx')
df.head()
```

```python
import pandas as pd

# Create a new column 'Combined Coach' by combining values from the specified columns
df['Combined Coach'] = df['Head coach[3]'].combine_first(df['Head coach']).combine_first(df['Head Coach2']).combine_first(df['First Team Manager']).combine_first(df['Manager'])

# Drop the original columns if needed
df = df.drop(['Head coach[3]', 'Head coach', 'Head Coach2', 'First Team Manager', 'Manager'], axis=1)

# Display the updated DataFrame
df.head()
```

```python
import pandas as pd

# Create a new column 'Combined Stadium' by combining values from the specified columns
df['Combined Stadium'] = df['Stadium'].combine_first(df['Ground'])

# Drop the original columns if needed
df = df.drop(['Stadium', 'Ground'], axis=1)

# Display the updated DataFrame
df.head()
```

```python
# Convert 'Capacity' column to numeric, handling errors with coerce
df['Capacity'] = pd.to_numeric(df['Capacity'], errors='coerce')

# Fill NaN values with 0 or any other suitable value
df['Capacity'].fillna(0, inplace=True)

# Convert the 'Capacity' column to integer
df['Capacity'] = df['Capacity'].astype(int)

# Display the updated Data
df.head()
```

```python
df = df.drop(['Chairman'], axis =1)
df.head()
```

```python
df = df.drop(['Owner'], axis =1)
df.head()
```

```python
# Finding capacity values that did not get handeld the rightway
# Setting a threshold for identifying outliers
threshold = 100000

# Identify clubs with capacity above the threshold or negative capacity
outliers = df[(df['Capacity'] > threshold) | (df['Capacity'] < 0)][['Club', 'Capacity']]

print("Clubs with high or negative capacity:")
print(outliers)

```

```python
# Now we will manually replace these values with the correct values
# Birmingham: 29,409, # Bradford 3.500 # Glossop: 1,301
correct_values = {
    'Birmingham City': 29409,
    'Bradford Park Avenue': 3500,
    'Glossop North End': 1301
}

# Update incorrect capacity values with correct ones
for club, correct_value in correct_values.items():
    df.loc[df['Club'] == club, 'Capacity'] = correct_value

print("DataFrame after updating incorrect capacity values:")
print(df)
```
